```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy import stats
from scipy.stats import linregress
from pprint import pprint

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

#creating a base url to use for city weather api calls
base_url = " http://api.openweathermap.org/data/2.5/weather"
#storing units as metric to use with the api calls
units = "imperial"

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
lat = []
lng = []
city_id = []
country = []
humidity = []
cloudiness = []
city_temps = []
wind_speed = []
date = []
min_temp = []
max_temp = []
city_name = []


for city in cities:
    
    try:
        query = f"{base_url}?q={city}&appid={weather_api_key}&units={units}"
        response = requests.get(query).json()
        print(f"The weather in {city} id: {response['id']} is {response['main']['temp']}F")
        city_name.append(response['name'])
        lat.append(round(response['coord']['lat'],2))
        lng.append(round(response['coord']['lon'],2))
        city_id.append(response['id'])
        country.append(response['sys']['country'])
        humidity.append(response['main']['humidity'])
        cloudiness.append(response['clouds']['all'])
        city_temps.append(response['main']['temp'])
        wind_speed.append(response['wind']['speed'])
        min_temp.append(response['main']['temp_min'])
        max_temp.append(response['main']['temp_max'])
        date.append(response['dt'])
    except KeyError:
        print (f"The city {city} was not found")

```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
weather_dict = {
    "id": city_id,
    "city": city_name,
    "country": country,
    "lat": lat,
    "lng": lng,
    "date": date,
    "temp": city_temps,
    "max temp": max_temp,
    "humidity": humidity,
    "cloudiness": cloudiness,
    "wind speed": wind_speed
}
weather_data = pd.DataFrame(weather_dict)
weather_data.head()
```

```python
weather_data.to_csv(output_data_file, index = False)
```

## Linear Regression

```python
northern_hemisphere = weather_data[weather_data['lat']>=0]
northern_hemisphere.head()
```

```python
southern_hemisphere = weather_data[weather_data['lat']<0]
southern_hemisphere.head()
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x_values = northern_hemisphere['lat']
y_values = northern_hemisphere['max temp']

(slope, intercept, rvalue, pvalue, stderr) = stats.linregress(x_values, y_values)

# Get regression values
regress_values = x_values * slope + intercept
#print(regress_values)
# Create line equation string
line_eq = "y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))

plt.scatter(x_values,y_values,edgecolors="black",facecolors="skyblue")
plt.plot(x_values,regress_values,"r-")

plt.title("Northern Hemisphere Latitude vs. Max Temp 4/26/2021")
plt.xlabel('Latitude')
plt.ylabel('Max Temp')
plt.annotate(line_eq,(10,15),fontsize=15,color="red")
plt.grid()

plt.savefig("output_data/Northern Hemisphere Latitude vs. Max Temp.png")

plt.show()
```
